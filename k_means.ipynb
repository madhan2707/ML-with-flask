```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

file_path = '/content/income.csv'
df = pd.read_csv(file_path)

X = df[['Age', 'Income($)']]

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_predicted = kmeans.fit_predict(X)

df['cluster'] = y_predicted

plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Income($)'], c=df['cluster'], cmap='viridis', alpha=0.6, edgecolors='black')

centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red', label='Centroids')

plt.xlabel('Age', fontsize=12)
plt.ylabel('Income ($)', fontsize=12)
plt.title('K-Means Clustering: Age vs Income', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
# prompt: db scan

from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
from google.colab import drive
import matplotlib.pyplot as plt
%matplotlib inline

drive.mount('/content/drive')

file_path = "/content/income.csv"
df = pd.read_csv(file_path)

scaler = MinMaxScaler()
df[['Age', 'Income($)']] = scaler.fit_transform(df[['Age', 'Income($)']])

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_predicted = kmeans.fit_predict(df[['Age', 'Income($)']])
df['cluster'] = y_predicted

# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Income($)'], c=df['cluster'], cmap='viridis', alpha=0.6, edgecolors='black')
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red', label='Centroids')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Income ($)', fontsize=12)
plt.title('K-Means Clustering: Age vs Income', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler # Import StandardScaler

print("Columns in dataset:", df.columns)

df.columns = df.columns.str.strip()

corrected_columns = ['Fresh', 'Milk']  # Update based on actual columns
X = df[corrected_columns].values  # Convert to NumPy array

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

dbscan = DBSCAN(eps=0.5, min_samples=5)
y_predicted = dbscan.fit_predict(X_scaled)

df['cluster'] = y_predicted

plt.figure(figsize=(8, 6))
plt.scatter(df['Fresh'], df['Milk'], c=df['cluster'], cmap='viridis', alpha=0.6, edgecolors='black')

plt.xlabel('Fresh Product Spend', fontsize=12)
plt.ylabel('Milk Product Spend', fontsize=12)
plt.title('DBSCAN Clustering: Wholesale Customers', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
